```python
import pandas as pd
import random
import requests
import io
import hvplot.pandas
```

```python
 # Downloading the csv file from GitHub 

url = "https://raw.githubusercontent.com/brown016340/Project-1/development/Resources/2018_Squirrel_Data.csv" # Make sure the url is the raw version of the file on GitHub
download = requests.get(url).content

# Reading the downloaded content and turning it into a pandas dataframe

squirrel_df = pd.read_csv(io.StringIO(download.decode('utf-8')))
squirrel_df = squirrel_df.dropna(subset=['Primary Fur Color'])
```

```python
color_map = squirrel_df.hvplot.points(
    "X",
    "Y",
    geo = True,
    tiles = "ESRI",
    height=900,
    width=700,
    size = 15,
    alpha = .25,
    color = "Primary Fur Color",
    title = "Primary Fur Color",
    xlim=(-73.98, -73.95), 
    ylim=(40.765, 40.80)
)
color_map
```

```python
hvplot.save(color_map, 'color.png')
```
